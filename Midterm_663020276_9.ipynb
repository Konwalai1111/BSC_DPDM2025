````python
# Calculate the number of missing 'tele_station_id' values per tele_station_oldcode
missing_tele_station_id_counts = DailyRain_df_filtered.groupby('tele_station_oldcode')['tele_station_id'].apply(lambda x: x.isnull().sum())

# Calculate the total number of records per tele_station_oldcode
total_records_per_station = DailyRain_df_filtered.groupby('tele_station_oldcode').size()

# Calculate the proportion of missing 'tele_station_id'
missing_tele_station_id_proportion = (missing_tele_station_id_counts / total_records_per_station) * 100

# Create a DataFrame for plotting
plot_data = pd.DataFrame({
    'Station Code': missing_tele_station_id_proportion.index,
    'Missing Proportion (%)': missing_tele_station_id_proportion.values
})

# Filter out stations with 0% missing to focus on actual missing data, or keep all if desired
# plot_data = plot_data[plot_data['Missing Proportion (%)'] > 0]

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Station Code', y='Missing Proportion (%)', data=plot_data)
plt.title('Proportion of Missing tele_station_id per Station')
plt.xlabel('Station Code')
plt.ylabel('Missing Proportion (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("Missing tele_station_id proportion per station:")
print(plot_data)
````

## ข้อที่ 5 เติมข้อมูลที่หายไปด้วยค่า Median ตาม Day of Year จาก historical แยกตามสถานี ตรวจสอบว่าไม่มี record หายอีก

````python
DailyRain_df_filtered
````

````python
# Create a copy to avoid SettingWithCopyWarning
df_to_impute = DailyRain_df_filtered.copy()

# Calculate the median 'tele_station_id' for each 'tele_station_oldcode' group
median_tele_station_id_per_oldcode = df_to_impute.groupby('tele_station_oldcode')['tele_station_id'].median()

# Fill missing 'tele_station_id' values using the calculated medians
# Use transform to broadcast the median back to the original DataFrame's shape
df_to_impute['tele_station_id'] = df_to_impute.groupby('tele_station_oldcode')['tele_station_id'].transform(lambda x: x.fillna(x.median()))

# Verify if there are any missing 'tele_station_id' values left
missing_tele_station_id_after_imputation = df_to_impute['tele_station_id'].isnull().sum()

print(f"Number of missing 'tele_station_id' records after imputation: {missing_tele_station_id_after_imputation}")
if missing_tele_station_id_after_imputation == 0:
    print("All missing 'tele_station_id' records have been imputed.")
else:
    print("There are still missing 'tele_station_id' records after imputation.")

print("\nFirst 5 rows of the DataFrame after 'tele_station_id' imputation:")
print(df_to_impute.head())

# Update DailyRain_df_filtered to reflect the imputed values (if desired, or use df_to_impute directly)
DailyRain_df_median = df_to_impute
DailyRain_df_median
````

## ข้อที่ 6 อภิปราย: เหตุใดใช้ Median แทน Mean

1. ความทนทานต่อค่าผิดปกติ (Robustness to Outliers): ค่ามัธยฐานได้รับผลกระทบจากค่าผิดปกติ (outliers) น้อยกว่าค่าเฉลี่ย (Mean) หาก tele_station_id มีข้อมูลบางค่าที่ผิดปกติหรือมีความคลาดเคลื่อน ค่ามัธยฐานจะช่วยให้การเติมข้อมูลมีความน่าเชื่อถือและเป็นตัวแทนของข้อมูลส่วนใหญ่ได้ดีกว่า
2. รักษาการกระจายตัวของข้อมูล: สำหรับข้อมูลที่มีการกระจายตัวแบบเบ้ (Skewed Distribution) ค่ามัธยฐานจะสะท้อนค่ากลางของข้อมูลได้ดีกว่าค่าเฉลี่ย ซึ่งอาจจะถูกดึงไปในทิศทางของค่าที่เบ้

````python
# Ensure 'rainfall_datetime' is in datetime format
DailyRain_df_median['rainfall_datetime'] = pd.to_datetime(DailyRain_df_median['rainfall_datetime'])

# Extract Year and Month
DailyRain_df_median['Year'] = DailyRain_df_median['rainfall_datetime'].dt.year
DailyRain_df_median['Month'] = DailyRain_df_median['rainfall_datetime'].dt.month

# Calculate monthly rainfall totals for each station
monthly_rainfall = DailyRain_df_median.groupby(['tele_station_oldcode', 'Year', 'Month'])['rainfall_value'].sum().reset_index()

# Rename columns to match desired output format
monthly_rainfall = monthly_rainfall.rename(columns={
    'tele_station_oldcode': 'Node',
    'rainfall_value': 'rain'
})

# Corrected: Use 'Year' and 'Month' (uppercase) as they exist in the DataFrame
output_monthly_rainfall = monthly_rainfall[['Node', 'rain', 'Month', 'Year']]

print("Monthly Rainfall Totals (First 5 rows):")
print(output_monthly_rainfall.head())

print("\nShape of the monthly rainfall DataFrame:")
print(output_monthly_rainfall.shape)

output_monthly_rainfall
````
